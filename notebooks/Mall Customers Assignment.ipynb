# A mini project on analysing  Mall Customers database

```python
#importing seaborn
import seaborn as sns


#importing dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import IPython
from IPython.display import Image
import warnings
```

```python
mall = pd.read_csv('Mall_Customers.csv')
```

```python
mall.head()
```

```python
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (18, 8)

# plt.subplot(1,2,1)
sns.set(style = 'whitegrid')
sns.displot(mall['Annual_Income_(k$)'])
plt.title('Distribution of Annual Income', fontsize = 20)
plt.xlabel('Range of Annual Income')
plt.ylabel('Count')
plt.show()


# plt.subplot(1,2,2)
sns.set(style = 'whitegrid')
sns.displot(mall['Age'], color = 'red')
plt.title('Distribution of Age', fontsize = 20)
plt.xlabel('Range of Age')
plt.ylabel('Count')
plt.show()
```

```python
labels = ['Female', 'Male']
size = mall['Gender'].value_counts()
colors = ['lightgreen', 'orange']
explode = [0,0.1]

plt.rcParams['figure.figsize'] = (9,9)
plt.pie(size, colors = colors, explode = explode, labels = labels, shadow = True, autopct = '%.1f%%')
plt.title('Gender', fontsize = 20)
plt.axis('off')
plt.legend()
plt.savefig("Mall-Analysis/artifacts/gender.pdf", format='pdf')
plt.show()
```

```python
#Q1. Answer: The females are in the lead with a share of 56% whereas the males have a share of 44%
#This is a huge gap specially when the population of the males is higher than the females.

#Q2. Answer: Most of the age group earning higher are the ages ranging between 30 - 35years.
```

```python
plt.rcParams['figure.figsize'] = (20,8)
sns.countplot(mall['Spending_Score'], palette = 'copper')
plt.title('Distribution of Spending Score', fontsize = 20)
plt.savefig("Mall-Analysis/artifacts/distribution.pdf", format='pdf')
plt.show()
```

```python
#Q3. Ans: with the above summary, we may conclude that most of the spending age is in the range 35-60
```
